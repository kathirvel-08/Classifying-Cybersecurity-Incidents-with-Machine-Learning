# 1. Incident grade based categorical impact

```python
import matplotlib.pyplot as plt
for category in df['IncidentGrade'].dropna().unique():
    frame = df[df['IncidentGrade'] == category]
    frame['Category'].value_counts().plot(kind='barh')
    plt.title(f'{category} Impact Count')
    plt.show()
```

# 2.Entity type based on Incident Grade

```python
for category in df['IncidentGrade'].dropna().unique():
    frame = df[df['IncidentGrade'] == category]
    frame['EntityType'].value_counts().plot(kind='barh')
    plt.title(f'{category} Impact Count')
    plt.show()
```

# 3. Evidence Role as per Incident Grade

```python
for category in df['IncidentGrade'].dropna().unique():
    frame = df[df['IncidentGrade'] == category]
    frame['EvidenceRole'].value_counts().plot(kind='barh')
    plt.title(f'{category} Impact Count')
    plt.show()
```

```python
for category in df['IncidentGrade'].dropna().unique():
    frame = df[df['IncidentGrade'] == category]
    frame['Year'].value_counts().plot(kind='barh')
    plt.title(f'{category} Impact Count')
    plt.show()
```
