```python
import pandas as pd
df=pd.read_csv("E:/Project/Project files/Microsoft/Data/new_train_sample.csv")
df.info()
```

```python
#Null Values
Null_columns = df.isnull().sum().reset_index()

null_list = []
for col in Null_columns[Null_columns[0]>0]['index']:
    null_list.append(col)
null_list.remove('IncidentGrade')

#drop null columns
df.drop(columns=null_list,axis=1,inplace=True)
```

```python
df.columns
```

```python
df.drop(columns='Unnamed: 0',axis=1,inplace=True)
```

```python
#remove duplicates
df.drop_duplicates(inplace=True)
```

```python
df.shape
```

```python
#Find unique values
unique_dict={"Col":[],"Unique":[]}
for col in df.columns:
    unique_dict['Col'].append(col)
    unique_dict['Unique'].append(df[col].nunique())
unique_df = pd.DataFrame(unique_dict).sort_values(by=['Unique'],ascending=False)
unique_df
```

```python
# id columns
id_col=[]
for col in df.columns:
    if 'Id' in col:
        id_col.append(col)
print(id_col)
```

```python
#remove id columns
df.drop(columns=id_col,axis=1,inplace=True)
```

```python
df.drop(columns="AccountSid",inplace=True,axis=1)
```

```python
#Find unique values
unique_dict={"Col":[],"Unique":[]}
for col in df.columns:
    unique_dict['Col'].append(col)
    unique_dict['Unique'].append(df[col].nunique())
unique_df = pd.DataFrame(unique_dict).sort_values(by=['Unique'],ascending=False)
unique_df
```

```python
df['Timestamp']=pd.to_datetime(df['Timestamp'])
```

```python
df['Year']=df['Timestamp'].dt.year
df['Month']=df['Timestamp'].dt.month
df['Day']=df['Timestamp'].dt.day
```

```python
#Timestamp
df.drop(columns="Timestamp",axis=1,inplace=True)
```
